```python
# importing required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef,
    confusion_matrix, roc_curve, auc,
)
from sklearn.svm import SVC
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
# Data Collection
df = pd.read_csv("CGU_Dataset-6.csv")
# Display the first few rows of the dataset to inspect its structure and content.
print("First 5 rows of the dataset:-\n", df.head())
```

# Data Preprocessing

```python
# Check the dimensions of the dataset (number of rows and columns).
row, col = df.shape
print("No. of rows in the dataset: ", row)
print("No. of column in the dataset: ", col)
```

```python
# Identify the data types of each column (numeric, categorical, text, etc.).
print("Data types of each column:\n", df.dtypes)
```

```python
# Display the number of missing values in each column
missingValues = df.isnull().sum()
print("Missing values per column:-")
print(missingValues)
```

```python
# Finding Unique categories of class column
print("Types of Class: ", df['CLASS'].unique())
```

```python
# Mapping with integer values
df['CLASS'] = df['CLASS'].map({'N': 0, 'Y': 1})
print("Checking Dataset after mapping:-\n", df.tail())
```

```python
# Split the dataset into independent and dependent feature
features = df.drop('CLASS', axis=1)
target = df.CLASS
```

```python
# Checking for NaN values
features.columns[features.isna().any()]
```

```python
# Standardize the dataset using StandardScaler
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
# Split the dataset into training and testing sets(70% training, 30%testing)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
# Display the size of the training and testing sets
print(f'Training set size: {X_train.shape[0]} samples \nTest set size: {X_test.shape[0]} samples')
```

# *Creation of SVM models using different kernels.*
# WITHOUT FEATURE DIMENSIONALITY REDUCTION

```python
# Define SVM models
svm_models = {
    "SVM (Linear)": SVC(kernel='linear', probability=True),
    "SVM (Polynomial)": SVC(kernel='poly', degree=2, probability=True),
    "SVM (RBF)": SVC(kernel='rbf', probability=True)
}
```

```python
# Function to calculate specificity
def specificity_score(y_true, y_pred):
    tn, fp, _, _ = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0.0
    return specificity
```

```python
# Create a table to store the results
results_table = {
    "Model": list(svm_models.keys()),
    "Accuracy": [],
    "Sensitivity": [],
    "Specificity": [],
    "Precision": [],
    "F1 Score": [],
    "MCC": []
}
```

```python
# Evaluate each SVM model and populate the results table
for model_name, model in svm_models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Evaluate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    sensitivity = recall_score(y_test, y_pred)
    specificity = specificity_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)    # Handle zero division
    f1 = f1_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    # Fill in the results table
    results_table["Accuracy"].append(accuracy)
    results_table["Sensitivity"].append(sensitivity)
    results_table["Specificity"].append(specificity)
    results_table["Precision"].append(precision)
    results_table["F1 Score"].append(f1)
    results_table["MCC"].append(mcc)

# Create a DataFrame from the results table
df_results = pd.DataFrame(results_table)

# Display the table
print(tabulate(df_results, headers='keys', tablefmt='pretty', showindex=False))
```

```python
# Find the index of the row with the highest accuracy
best_index = df_results["Accuracy"].idxmax()

# Extract the information about the best classifier
best_classifier_name = df_results.loc[best_index, "Model"]
best_classifier_accuracy = df_results.loc[best_index, "Accuracy"]

# Print conclusion
print(f"The best SVM classifier is '{best_classifier_name}' with an accuracy of {best_classifier_accuracy:.2%}.")
```

```python
# Access the best classifier
best_classifier = svm_models[best_classifier_name]

# Train the best classifier
best_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_classifier.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix for the best SVM classifier
sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title(f"Confusion Matrix - {best_classifier_name}")
plt.show()
```

```python
# Plot ROC curves for SVM models
for name, clf in svm_models.items():
    clf.fit(X_train, y_train)
    y_score = clf.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot the diagonal line
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves - SVM Kernels')
plt.legend(fontsize='small')
plt.show()
```

# WITH FEATURE DIMENSIONALITY REDUCTION TECHNIQUE (PCA)

```python
# Apply PCA for dimensionality reduction
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Scatter plot for the original dataset and dataset after applying PCA
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot for the original dataset
axs[0].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', marker='o', edgecolors='k')
axs[0].set_title('Scatter Plot of Original Dataset')
axs[0].set_xlabel('Feature 1')
axs[0].set_ylabel('Feature 2')

# Scatter plot for the dataset after applying PCA
axs[1].scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', marker='o', edgecolors='k')
axs[1].set_title('Scatter Plot of Dataset after PCA')
axs[1].set_xlabel('Principal Component 1')
axs[1].set_ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

```python
# Create a table to store the results with PCA
results_table_pca = {
    "Model": [],  # Combined Model and PCA
    "Num Features": [],  # New column for Number of Features
    "Accuracy": [],
    "Sensitivity": [],
    "Specificity": [],
    "Precision": [],
    "F1 Score": [],
    "MCC": []
}
```

```python
# Evaluate each SVM model with PCA and populate the results table
for model_name, model in svm_models.items():
    # Train the SVM model with PCA
    model.fit(X_train_pca, y_train)

    # Make predictions on the test set with PCA
    y_pred_pca_svm = model.predict(X_test_pca)

    # Evaluate performance metrics with PCA
    accuracy_pca_svm = accuracy_score(y_test, y_pred_pca_svm)
    sensitivity_pca_svm = recall_score(y_test, y_pred_pca_svm)
    specificity_pca_svm = specificity_score(y_test, y_pred_pca_svm)
    precision_pca_svm = precision_score(y_test, y_pred_pca_svm, zero_division=0)    # Handle zero division
    f1_pca_svm = f1_score(y_test, y_pred_pca_svm)
    mcc_pca_svm = matthews_corrcoef(y_test, y_pred_pca_svm)

    # Fill in the results table with PCA for SVM
    results_table_pca["Model"].append(f'PCA + {model_name}')
    results_table_pca["Num Features"].append(X_train_pca.shape[1])  # Number of PCA features
    results_table_pca["Accuracy"].append(accuracy_pca_svm)
    results_table_pca["Sensitivity"].append(sensitivity_pca_svm)
    results_table_pca["Specificity"].append(specificity_pca_svm)
    results_table_pca["Precision"].append(precision_pca_svm)
    results_table_pca["F1 Score"].append(f1_pca_svm)
    results_table_pca["MCC"].append(mcc_pca_svm)

# Create a DataFrame from the results table with PCA for SVM
df_results_pca_svm = pd.DataFrame(results_table_pca)

# Display the table with PCA for SVM
print(tabulate(df_results_pca_svm, headers='keys', tablefmt='pretty', showindex=False))
```

```python
# Find the index of the row with the highest accuracy with PCA for SVM
best_index_pca_svm = df_results_pca_svm["Accuracy"].idxmax()

# Extract the information about the best SVM classifier with PCA
best_classifier_name_pca = df_results_pca_svm.loc[best_index_pca_svm, "Model"]
best_classifier_accuracy_pca = df_results_pca_svm.loc[best_index_pca_svm, "Accuracy"]

# Print conclusion with PCA for SVM
print(f"The best SVM classifier with PCA is '{best_classifier_name_pca}' with an accuracy of {best_classifier_accuracy_pca:.2%}.")
```

```python
# Access the best SVM classifier with PCA
best_classifier_pca_svm = svm_models[best_classifier_name_pca.split('PCA + ')[-1]]

# Train the best SVM classifier with PCA
best_classifier_pca_svm.fit(X_train_pca, y_train)

# Make predictions on the test set with PCA for SVM
y_pred_pca_svm = best_classifier_pca_svm.predict(X_test_pca)

# Compute confusion matrix with PCA for SVM
cm_pca_svm = confusion_matrix(y_test, y_pred_pca_svm)

# Plot confusion matrix for the best SVM classifier with PCA
sns.set(font_scale=1.2)
sns.heatmap(cm_pca_svm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title(f"Confusion Matrix - {best_classifier_name_pca}")
plt.show()
```

```python
# Plot ROC curves for SVM models with PCA
for name, clf in svm_models.items():
    clf.fit(X_train_pca, y_train)
    y_score_pca_svm = clf.predict_proba(X_test_pca)[:, 1]
    fpr_pca_svm, tpr_pca_svm, _ = roc_curve(y_test, y_score_pca_svm)
    roc_auc_pca_svm = auc(fpr_pca_svm, tpr_pca_svm)
    plt.plot(fpr_pca_svm, tpr_pca_svm, label=f'{name} (AUC = {roc_auc_pca_svm:.2f})')

# Plot the diagonal line for PCA for SVM
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves with PCA for SVM')
plt.legend(fontsize='x-small')
plt.show()
```

```python
# Bar plot for accuracy comparison
plt.figure(figsize=(10, 6))
bars = plt.bar(["Without PCA", "With PCA"], [best_classifier_accuracy, best_classifier_accuracy_pca], color=['blue', 'green'])

# Add values inside the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{round(yval * 100, 2)}%", ha='center', va='bottom')

plt.ylabel("Accuracy")
plt.title("Accuracy Comparison: With Bar plot")
plt.show()
```
